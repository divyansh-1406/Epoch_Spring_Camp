```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.array([
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
])
data[:, 3][data[:, 3] == 'Apple'] = 0
data[:, 3][data[:, 3] == 'Banana'] = 1
data[:, 3][data[:, 3] == 'Orange'] = 2
```

```python
data
```

```python
X = data[: , 0:3].astype(float)
y = data[: , 3].astype(int)
```

```python
def euclideanDistance(a , b):
    k = np.array(a , dtype = float) - np.array(b , dtype = float)
    t = np.dot(k,k)
    return t**0.5
```

```python
def Manhattan(a,b):
    dist = np.array(a , dtype = float) - np.array(b , dtype = float)
    dist = abs(dist)
    return np.sum(dist)
```

```python
def Minkowski(a,b,p):
    dist = np.array(a , dtype = float) - np.array(b , dtype = float)
    dist = dist ** p
    t = np.sum(dist)
    return t**(1/p)
```

```python
class KNN:
    def __init__(self,k=3):
        self.k = k
    def fit(self,X,y):
        self.X = X
        self.y = y
        
    def predict_one(self,x):
        X_dist_pred = np.array([[0]*2] * len(self.y), dtype=float)
        for i in range (len(self.y)):
            X_dist_pred[i][0] , X_dist_pred[i][1] = euclideanDistance(x , self.X[i] ) , self.y[i] #storing euclidean distance along with probable output value
        
        
        sorted_X =X_dist_pred [X_dist_pred[:, 0].argsort()] #sorting as per the Euclidean Distance
        topkvalues = sorted_X[:self.k , 1] #choosing top k values based on the smallest k euclidean distance
        #calculating what fruit value appears the most
        value , count = np.unique(topkvalues , return_counts = True) 
        maxcount = np.argmax(count)
        return value[maxcount]
        
    def predict(self,X_pred):
        y_pred = np.zeros(X_pred.shape[0], dtype=int)
        for i in range(X_pred.shape[0]):
            y_pred[i] = self.predict_one(X_pred[i])
        return y_pred
    
```

TESTING ON TEST DATA

```python
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
] , dtype = float)
model = KNN(k=3 ) 
model.fit(X,y)
predictions = model.predict(test_data)
predictions = ["Apple" if x == 0 else x for x in predictions]
predictions = ["Banana" if x == 1 else x for x in predictions]
predictions = ["Orange" if x == 2 else x for x in predictions]
print(predictions)
true_label = ['Banana' , 'Apple' , 'Orange']
```

```python
#Checking with different values of k
model = KNN(k=1)
model.fit(X,y)
predictions = model.predict(test_data)
predictions = ["Apple" if x == 0 else x for x in predictions]
predictions = ["Banana" if x == 1 else x for x in predictions]
predictions = ["Orange" if x == 2 else x for x in predictions]
print(predictions)
```

```python
#trying normalization 
X_normalized = (X - X.mean(axis=0)) / X.std(axis=0)

model.fit(X_normalized,y)
predictions = model.predict(test_data)
predictions = ["Apple" if x == 0 else x for x in predictions]
predictions = ["Banana" if x == 1 else x for x in predictions]
predictions = ["Orange" if x == 2 else x for x in predictions]
print(predictions)
```

```python
def accuracy( true_labels , predictions ) :
    c = 0
    for i in range(len(true_labels)):
        if true_labels[i] == predictions[i] :
            c+=1 
    return (c/len(predictions))*100
```

```python
condition = round(len(X)*0.7)
train_X , test_X , train_y , test_y = X[0:condition] , X[condition:] , y[0:condition] , y[condition:] 
model.fit(train_X , train_y)
predictions = model.predict(test_X)
print (accuracy (test_y , predictions))
```
